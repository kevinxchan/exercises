Given a string s, find the longest palindromic substring in s. You may assume that the maximum length of s is 1000.

Example 1:

<code>
Input: "babad"
Output: "bab"
</code>

Note: "aba" is also a valid answer.

Example 2:

<code>
Input: "cbbd"
Output: "bb"
</code>

```python
def getCharAtIndex(string, index):
    if index < 0 or index >= len(string):
        return ""
    else:
        return string[index] 
```

```python
def longestPalindrome(s):
    longest = ""
    n = len(s)
    idx, idxRight, idxLeft = 0, 0, 0
    subStr = ""
    restart = True
    while idx < n: 
        if restart:
            idxRight = idx + 1
            idxLeft = idx - 1
        
        curr = s[idx]
        currRight = getCharAtIndex(s, idxRight)
        currLeft = getCharAtIndex(s, idxLeft)
        
        if currRight == currLeft:
            subStr = s[idxLeft:idxRight + 1] # doesn't include :RIGHT index
            if idxRight < n and idxLeft > 0:
                idxRight += 1
                idxLeft -= 1
                restart = False
                
            else:
                idx += 1
                restart = True
        else:
            idx += 1
            restart = True
        
        if len(longest) < len(subStr):
            longest = subStr
    return longest
```

```python
print longestPalindrome("cbbd")
```
